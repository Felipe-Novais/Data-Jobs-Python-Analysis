````python
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
````

## Exporting data to files

```
.to_csv()
.to_excel()
```

````python
# create copy of original dataframe
quarter_1_df = df.copy()

# create month column for data filtering
quarter_1_df["job_posted_month"] = quarter_1_df["job_posted_date"].dt.month

# get months from first quarter
q1_months = [1, 2, 3]

# filter the dataframe to have only the requested data and final cleanups
quarter_1_df = quarter_1_df.loc[quarter_1_df["job_posted_month"].isin(q1_months)]
quarter_1_df.reset_index(drop=True, inplace=True)
quarter_1_df.drop(columns="job_posted_month", inplace=True)

quarter_1_df.head(3)
````

###  Example 1: extract to csv

````python
quarter_1_df.to_csv("first_quarter_data.csv")

csv_creation_test = pd.read_csv("first_quarter_data.csv", index_col=0)
````
