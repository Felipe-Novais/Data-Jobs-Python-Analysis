```python
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

## Concat Example 1 (SQL Union)

```python
# example data
job_postings_jan = pd.DataFrame({
    "job_id":[1, 2, 3, 4, 5],
    "job_title":["Data Scientist", "Data Analyst", "Machine Learning Engineer", "Data Scientist", "Data Engineer"],
    "company":["Company A", "Company B", "Company C", "Company D", "Company E"],
    "job_posted_date":["2024-01-02", "2024-01-07", "2024-01-14", "2024-01-19", "2024-01-24"]
})

job_postings_feb = pd.DataFrame({
    "job_id":[6, 7, 8, 9, 10],
    "job_title":["Data Scientist", "Data Analyst", "Machine Learning Engineer", "Data Scientist", "Data Engineer"],
    "company":["Company F", "Company G", "Company H", "Company I", "Company J"],
    "job_posted_date":["2024-02-05", "2024-02-09", "2024-02-12", "2024-02-18", "2024-02-22"]
})
```

```python
# example 1
pd.concat(objs=[job_postings_jan, job_postings_feb], ignore_index=True)
```

```python
# example 2
example_df = pd.concat(objs=[job_postings_jan, job_postings_feb]).set_index("job_id")
example_df["job_posted_date"] = pd.to_datetime(example_df["job_posted_date"])

example_df
```

## Concat Example 2 (SQL Union)

```python
df.info()
```

```python
# separating dataframes based on months

df["job_posted_month"] = df["job_posted_date"].dt.strftime("%b")
months = df["job_posted_month"].unique()

dict_months = {month:df.loc[df["job_posted_month"] == month, :] for month in months}
```

```python
# retrieving data from 2023 first quarter

first_quarter_df = pd.concat(objs=[dict_months["Jan"], dict_months["Feb"], dict_months["Mar"]]).reset_index(drop=True)
first_quarter_df.drop(columns="job_posted_month", inplace=True)

first_quarter_df.head()
```
