```python
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# convert date column do datetime
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# iloc usage example

df.iloc[
    0:8,
    0:6
]
```

```python
# loc usage example 1

df.loc[
    :,
    ["job_title_short", "company_name", "job_country", "salary_year_avg"]
]
```

```python
# loc usage example 2

df.loc[
    df["salary_year_avg"].notna(),
    ["job_title_short", "company_name", "job_country", "salary_year_avg"]
].reset_index(drop=True)
```

```python
# loc usage example 3

df.loc[
    :,
    ["job_title_short", "company_name", "job_country", "salary_year_avg"]
].dropna(subset="salary_year_avg").sort_values(by="salary_year_avg", ascending=False).reset_index(drop=True)
```

```python
# loc usage example 4

df.loc[
    (df["salary_year_avg"].notna()) & (df["job_country"] == "Brazil"),
    ["job_title_short", "company_name", "job_country", "salary_year_avg"]
].sort_values(by="salary_year_avg", ascending=False).reset_index(drop=True)
```

```python
# loc usage example 5

df.loc[
    (df["salary_year_avg"].notna()) & (df["job_country"] == "India") & (df["job_title_short"].str.contains("Senior")),
    ["job_title_short", "company_name", "job_country", "salary_year_avg"]
].sort_values(by="salary_year_avg", ascending=False).reset_index(drop=True)
```
