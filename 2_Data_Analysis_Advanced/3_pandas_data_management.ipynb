```python
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# the .head() method does not return a true sample of the data

df.head(3)
```

```python
# the .sample() method is much better, as it provides random rows from the dataframe for better analysis

df.sample(3)
```

```python
# utilize random_state to return always the same sample result

df.sample(3, random_state=42)
```

```python
# to create copies of dataframes, it is better to utilize the .copy() method, as the equal sign does not create a new dataframe

df_test_1 = df

print(id(df))
print(id(df_test_1))
print("are the IDs equal?", "Yes" if id(df) == id(df_test_1) else "No")
```

```python
# .copy() method example

df_test_2 = df.copy()

print(id(df))
print(id(df_test_2))
print("are the IDs equal?", "Yes" if id(df) == id(df_test_2) else "No")
```
