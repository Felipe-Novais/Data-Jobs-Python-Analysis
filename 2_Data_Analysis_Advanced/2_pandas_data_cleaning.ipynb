```python
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
print(f"Unaltered dataframe total rows: {len(df)}")
```

```python
# data cleaning step 1: remove all around duplicates

df_duplicates_removed_1 = df.drop_duplicates()

print(f"Unaltered dataframe total rows: {len(df)}")
print(f"Duplicates removed dataframe total rows: {len(df_duplicates_removed_1)}")
print(f"Total duplicates removed: {len(df) - len(df_duplicates_removed_1)}")
```

```python
# data cleaning step 2: remove duplicates based on a subset, repeated job titles from the same company

df_duplicates_removed_2 = df.drop_duplicates(subset=["job_title", "company_name"])

print(f"Unaltered dataframe total rows: {len(df)}")
print(f"Duplicates removed dataframe total rows: {len(df_duplicates_removed_2)}")
print(f"Total duplicates removed: {len(df) - len(df_duplicates_removed_2)}")
```

```python
# data cleaning step 3: replace NaN values from salary columns with their median counterparts

median_yearly_salary = df["salary_year_avg"].median()
median_hourly_salary = df["salary_hour_avg"].median()

df_duplicates_removed_2.loc[:, ["salary_year_avg", "salary_hour_avg"]].head()
```

```python
df_fillna = df_duplicates_removed_2

df_fillna["salary_year_avg"] = df_duplicates_removed_2["salary_year_avg"].fillna(median_yearly_salary)
df_fillna["salary_hour_avg"] = df_duplicates_removed_2["salary_hour_avg"].fillna(median_hourly_salary)

df_fillna.head()
```
