```python
import pandas as pd
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

## Indexing Methods

df.index\
set_index()\
sort_index()\
reset_index()

```python
# recovering the indexes from a dataframe

df.index
```

```python
# data frame does not have an index name

df.sample(1)
```

```python
df.index.name = "job_index"
df.sample(1)
```

```python
# reset_index - resets the indexes of the dataframe and creates a new column with the old indexes

job_postings_brazil = df.loc[
    df["job_country"] == "Brazil",
    :
].reset_index()

job_postings_brazil.rename(columns={"job_index":"job_index_old"}, inplace=True)
job_postings_brazil.index.name = "job_index"

job_postings_brazil.sample(1)
```

```python
# set_index() - replaces the current indexes with a column from the dataframe

job_postings_brazil.set_index("job_index_old", inplace=True)
job_postings_brazil.index.name = "job_index"

job_postings_brazil.head(3)
```

```python
# sort_index() - sorts the dataframe rows based on the indexes

yearly_salary_analysis = df.pivot_table(index="job_title_short", values="salary_year_avg", aggfunc=["median", "min", "max"])
yearly_salary_analysis.sort_index(inplace=True)

yearly_salary_analysis
```

```python
yearly_salary_analysis.sort_values(by=("median", "salary_year_avg"), ascending=False)
```
