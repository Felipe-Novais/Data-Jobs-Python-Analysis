```python
import pandas as pd
from datasets import load_dataset

# data load
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# data clean
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# Filtering data from salary_year_avg column

df.loc[
    df["salary_year_avg"].notna(),
    "salary_year_avg"
]
```

## .apply() method - utilized to apply functions to rows and columns of a DataFrame

```python
# basic test without the .apply() method

salaries = df.loc[
    df["salary_year_avg"].notna(),
    ["job_title_short", "company_name","salary_year_avg"]
].reset_index(drop=True).copy()

salaries["adjusted_salary_1"] = salaries["salary_year_avg"] * 1.03 # inflation

salaries.head()
```

```python
# creating a simple function to utilize the .apply() method

def salary_inflation(salary: float) -> float:
    return salary * 1.03

salaries["adjusted_salary_2"] = salaries["salary_year_avg"].apply(salary_inflation)

salaries.head()
```

```python
# reseting the second column and utilizing a lambda function instead to create it

salaries.drop(columns=["adjusted_salary_2"], inplace=True)
```

```python
salaries["adjusted_salary_2"] = salaries["salary_year_avg"].apply(lambda salary: salary * 1.03)
salaries.head()
```

```python
import ast

list_string_test = "['Felipe', 'Maria', 'Gustavo']"
list_string_converted = ast.literal_eval(list_string_test)
print(list_string_converted, type(list_string_converted))
```

```python
def convert_str_to_list(job_skills: str) -> list:
    if pd.isna(job_skills):
        return job_skills
    else:
        return ast.literal_eval(job_skills)
    
df["job_skills"] = df["job_skills"].apply(convert_str_to_list)
```

## Utilizing the .apply() method on rows with axis=1

```python
# specifying the column in not necessary because the function will be applied on a row level, not column

def salary_inflation_2(row: pd.Series) -> pd.Series:
    if "Senior" in row["job_title_short"]:
        return row["salary_year_avg"] * 1.05
    else:
        return row["salary_year_avg"] * 1.03
    
salaries["adjusted_salary_3"] = salaries.apply(salary_inflation_2, axis=1)

salaries.loc[
    salaries["job_title_short"].str.contains("Senior"),
    :
]
```
