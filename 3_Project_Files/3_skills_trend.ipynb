```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from datasets import load_dataset

# load data
ds = load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda skill: ast.literal_eval(skill) if pd.notna(skill) else skill)
```

## Trending in demand skills for Data Analysts in the US

1 - Filter for US job postings and create job posting month column

```python
df_da_us = df.loc[
    (df["job_country"] == "United States") & (df["job_title_short"] == "Data Analyst"),
    :
]

df_da_us["job_posted_month"] = df_da_us["job_posted_date"].dt.month
```

2 - Explode the dataframe

```python
df_explode = df_da_us.explode("job_skills")
```

3 - Aggregation and sorting

```python
df_aggregation = df_explode.pivot_table(index="job_posted_month", columns="job_skills", aggfunc="size", fill_value=0)
df_aggregation
```

```python
df_aggregation.loc["Total"] = df_aggregation.sum()
df_aggregation
```

```python
top_5_skills = df_aggregation.sort_values("Total", axis=1, ascending=False).iloc[
    :,
    :5
]
```

```python
top_5_skills.drop("Total", inplace=True)
top_5_skills
```

4 - Calculating percentage and plotting

```python
job_postings_count = df_da_us["job_posted_month"].value_counts()
```

```python
import calendar

plot_df = top_5_skills.div(job_postings_count / 100, axis=0)
plot_df.index = [calendar.month_abbr[month] for month in plot_df.index]

plot_df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# plotting
sns.set_theme(style="ticks")
sns.lineplot(
    data=plot_df,
    dashes=False,
    marker="o",
)

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f"{int(y)}%"))
ax.legend(title="")
ax.set_ylim(15, 60)
ax.set_ylabel("Likelihood in Job Posting")
ax.set_xlabel("2023")
ax.set_title("Top 5 skills for Data Analysts in the US")
ax.grid(
    True,
    linestyle="--",
    alpha=0.6
)
```
