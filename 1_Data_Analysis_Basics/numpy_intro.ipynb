1 - Create a list of 1_000_000 salaries ranging from 50_000 to 150_000

```python
# utilizing "pure" Python

import random

salaries = [random.randint(50_000, 150_000) for x in range(1_000_000)]

print(len(salaries))
print(type(salaries))
```

```python
# utilizing numpy

import numpy as np

salaries = np.random.randint(50_000, 150_000, 1_000_000)

print(len(salaries))
print(type(salaries))
```

2 - Calculation examples

```python
# Job titles
job_titles = np.array(["Data Analyst", "Data Scientist", "Data Engineer", "Machine Learning Engineer"])

# Base salaries
base_salaries = np.array([60_000, 80_000, 75_000, 90_000])

# Bonus rates
bonus_rates = np.array([0.05, 0.1, 0.08, 0.12])
```

```python
total_salaries = base_salaries + (base_salaries * bonus_rates)
total_salaries.mean()
```

```python
# dealing with Null values, nan in numpy

job_titles = np.array(["Data Analyst", "Data Scientist", "Data Engineer", "Machine Learning Engineer", "AI Engineer"])
base_salaries = np.array([60_000, 80_000, 75_000, 90_000, np.nan])
bonus_rates = np.array([0.05, 0.1, 0.08, 0.12, np.nan])
```

```python
total_salaries = base_salaries + (base_salaries * bonus_rates)
np.nanmean(total_salaries)
```
