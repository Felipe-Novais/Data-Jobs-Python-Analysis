# 3 - Filtering data

```python
filtered_df_1 = df[["job_title_short", "job_location", "salary_year_avg"]][(df["job_title_short"] == "Data Analyst") & (df["salary_year_avg"].notna())]
filtered_df_1
```

```python
filtered_df_2 = df.loc[
    (df["job_title_short"] == "Data Analyst") & (df["salary_year_avg"].notna()),
    ["job_title_short", "job_location", "salary_year_avg"]
]
filtered_df_2
```
