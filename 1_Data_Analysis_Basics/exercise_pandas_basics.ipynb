```python
import pandas as pd
import datasets

# load dataframe
ds = datasets.load_dataset("lukebarousse/data_jobs")
df = ds["train"].to_pandas()

# alter type to datetime
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# solution 1

cleaned_df = df.loc[
    (df["job_country"] == "United States") & (df["salary_year_avg"].notna()),
    :
]

cleaned_df.groupby("job_title_short").agg(
    median_salary=("salary_year_avg", "median"),
    min_salary=("salary_year_avg", "min"),
    max_salary=("salary_year_avg", "max"),
    total_job_postings=("job_title_short", "count")
).sort_values(by="median_salary", ascending=False).reset_index()
```

```python
# solution 2

cleaned_df.groupby("job_title_short")["salary_year_avg"].agg(["median", "min", "max", "count"]).sort_values(by="median", ascending=False).reset_index()
```
