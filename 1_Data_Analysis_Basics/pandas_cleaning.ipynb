# 1 - Datetime conversion and new column creation

```python
# check type of job_posted_date column
type(df.loc[0, "job_posted_date"])
```

```python
# convert job_posted_date column from str to datetime
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# utilize dt parameter to add new column to the dataset
df["job_posted_month"] = df["job_posted_date"].dt.month
```

```python
# check new column added
df.loc[:,["job_title_short", "job_posted_date", "job_posted_month"]]
```

# 2 - Order by utilizing sort_values method

```python
# just sorting, selecting columns and printing the result
df.sort_values(by="job_posted_date", ascending=True).loc[
    :,
    ["job_title_short", "job_posted_date"]
]
```

```python
# sorting and also applying the sort to the DataFrame itself utilizing the inplace named argument
df.sort_values(by="job_posted_date", ascending=True, inplace=True)
```

```python
df.head(2)
```

# 3 - Drop columns with labels and axis named parameters, and NaN rows based on subset and axis parameters

```python
# drop the salary_hour_avg column
df.drop(labels="salary_hour_avg", axis=1, inplace=True)
```

```python
# drop the NaN salary_year_avg rows
df.dropna(subset=["salary_year_avg"], axis=0, inplace=True)
```
